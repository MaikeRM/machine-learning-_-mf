```python
import pandas as pd
import yfinance as yf
import seaborn as sns
```

```python
acoes = ['PETR4', 'ITUB4', 'JBSS3', 'CSNA3']
acoes = list(map(lambda x: x+'.SA', acoes))

'''for i in range(len(acoes)):
    acoes[i] = acoes[i] + '.SA'
'''
precos = yf.download(acoes, period='4y')['Adj Close']
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (15,30))

for i in range(len(acoes)):
    plt.subplot(len(acoes), 1, i + 1)
    sns.histplot(precos[acoes[i]], kde = True, bins=100)
```

```python
plt.figure(figsize = (15,20))

for i in range(len(acoes)):
    plt.subplot(len(acoes), 1, i + 1)
    sns.boxplot(precos[acoes[i]])
```

```python
ewma_50 = precos.ewm(alpha=0.01, min_periods=50, adjust=True).mean()
ewma_100 = precos.ewm(alpha=0.04, min_periods=50, adjust=True).mean()
```

```python
ewma_50['PETR4.SA'].plot()
ewma_100['PETR4.SA'].plot()
precos['PETR4.SA'].plot(figsize = (20,10))
```
